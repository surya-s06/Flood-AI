```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.DataFrame({

    "Version": [

        "Vision V1",
        "Vision V2",
        "Vision V3"

    ],

    "IoU": [

        0.5715,
        0.6227,
        0.6887

    ],

    "Dice Score": [

        0.7213,
        0.7629,
        0.8132

    ],

    "Pixel Accuracy": [

        0.8148,
        0.8280,
        0.8646

    ]

})

results
```

```python
results["IoU Improvement"] = (
    results["IoU"].diff()
)

results["Dice Improvement"] = (
    results["Dice Score"].diff()
)

results["Accuracy Improvement"] = (
    results["Pixel Accuracy"].diff()
)

results
```

```python
best = results.loc[
    results["IoU"].idxmax()
]

best
```
